# Winning Wrestlers Entertainment

In this activity you will be taking four seperate csvs that were scraped down from a wrestling database, merging them together, and then creating charts to visualize a wrestler's wins and losses over the course of four years.

### Part 1 - Macho Merging

* You will likely need to perform three different merges over the course of this activity, changing the names of your columns as you go along.

```python
# Import the necessary modules
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Bring each CSV into a separate data frame
WWE_2013 = pd.read_csv("../Resources/WWE-Data-2013.csv")
WWE_2014 = pd.read_csv("../Resources/WWE-Data-2014.csv")
WWE_2015 = pd.read_csv("../Resources/WWE-Data-2015.csv")
WWE_2016 = pd.read_csv("../Resources/WWE-Data-2016.csv")
WWE_2016.head(5)
```

```python
# Merge the first two datasets (2013 and 2014) on "Wrestler" so that no data is lost (should be 182 rows)\\
#print("2013 count")
#print(WWE_2013.count())

#print("2014 count")
#print(WWE_2014.count())
merge_1=pd.merge(WWE_2013, WWE_2014, on="Wrestler",how="outer")
merge_1.count()
```

```python
# Rename our _x columns to "2013 Wins", "2013 Losses", and "2013 Draws"
merge_1_rename=merge_1.rename(columns={"Wins_x":"2013 Wins","Losses_x":"2013 Losses","Draws_x":"2013 Draws",
                               "Wins_y":"2014 Wins","Losses_y":"2014 Losses","Draws_y":"2014 Draws" })
merge_1_rename.head()
# Rename our _y columns to "2014 Wins", "2014 Losses", and "2014 Draws"

```

```python
# Merge our newly combined dataframe with the 2015 dataframe
merge_2=pd.merge(merge_1_rename,WWE_2015,on="Wrestler",how="outer")
merge_2.head()
```

```python
# Rename "wins", "losses", and "draws" to "2015 Wins", "2015 Losses", and "2015 Draws"
merge_2_rename=merge_2.rename(columns={"Wins":"2015 Wins","Losses":"2015 Losses","Draws":"2015 Draws" })
merge_2_rename.head()
```

```python
# Merge our newly combined dataframe with the 2016 dataframe
merge_3=pd.merge(merge_2_rename,WWE_2016, on="Wrestler",how="outer")
merge_3
```

```python
# Rename "wins", "losses", and "draws" to "2016 Wins", "2016 Losses", and "2016 Draws"
merge_3_rename=merge_3.rename(columns={"Wins":"2016 Wins","Losses":"2016 Losses","Draws":"2016 Draws"})
merge_3_rename
```
